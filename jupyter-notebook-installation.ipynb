# Installation of Jupyter Notebook 

[Jupyter Notebook](https://jupyter.org/) က interactive program development နဲ့ data science နဲ့ပတ်သက်တဲ့ ပရောဂျက်တွေကို presentation လုပ်ဖို့အတွက်  
တကယ့်ကို အသုံးဝင်ပါတယ်။ Jupyter ကို installation လုပ်ဖို့အတွက်က Python version 3 နဲ့အထက် (သို့) Python 2.7 လိုအပ်ပါတယ်။   

## Installation of Anaconda

ကျွန်တော်ကတော့ [Anaconda](https://www.anaconda.com/download/#linux) ကို အရင် installation လုပ်ပြီးတော့၊ အဲဒီအပေါ်မှာ Python environment အသစ်တစ်ခု  
တည်ဆောက်ပြီးတော့မှ Jupyter ကို installation လုပ်ပါလို့ အကြံပေးချင်ပါတယ်။ ဘာကြောင့်လဲ ဆိုတော့ python library တွေကို installation လုပ်တဲ့ အခါမှာ၊ update လုပ်တဲ့အခါမှာ ကြုံလာရမယ့် library conflict ပြဿနာတွေကို ရှောင်နိုင်အောင်လို့ပါ။   

Anaconda installation နဲ့ ပတ်သတ်တဲ့ link:  
https://docs.anaconda.com/anaconda/install/linux

Jupyter installation နဲ့ ပတ်သက်တဲ့ link:  
http://jupyter.readthedocs.io/en/latest/install.html  

Ubuntu OS ပေါ်မှာ installation လုပ်တဲ့အဆင့်တွေကို အလွယ်ရှင်းပြရလျှင် အောက်ပါအတိုင်း ဖြစ်ပါတယ်။  

1. Anaconda ကို ဒီလင့်ကနေ (https://www.anaconda.com/download/#linux) Anaconda 5.2 (Python 3.6 version) ကို အရင်ဆုံး ဒေါင်းလုဒ်လုပ်ပါ။  
အဆင်ပြေပြေနဲ့ ဒေါင်းလုဒ် လုပ်တာပြီးသွားရင် ~/Download/ ဖိုလ်ဒါအောက်မှာ Anaconda3-5.2.0-Linux-x86_64.sh ဆိုတဲ့ဖိုင်  
ရှိပါလိမ့်မယ်။  

2. ဒေါင်းလုဒ်လုပ်ထားခဲ့တဲ့ ဖိုင်ကို data fingerprint အနေနဲ့သုံးတဲ့ သင်္ချာ function တစ်ခုဖြစ်တဲ့ SHA-256 (Secure Hash Algorithm, 256 bit) ကို သုံးထားတဲ့ checksum ပရိုဂရမ်တစ်ခုကို သုံးပြီး နဂိုဖိုင်နဲ့ တူမတူဆိုတာကို အောက်ပါအတိုင်း စစ်ဆေးနိုင်ပါတယ်။  
```
$ sha256sum Anaconda3-5.2.0-Linux-x86_64.sh  
```
အင်တာနက် ချိတ်ဆက်မှုက သိပ်မကောင်းတဲ့ကြားက ဒေါင်းလုဒ်လုပ်ခဲ့ရတယ် ဆိုရင်တော့ ဖိုင်ဆိုက်ကလည်း 622 MB ခန့်မို့လို့၊ ဒီအဆင့်ကို လုပ်သင့်ပါတယ်။  

3. Anaconda ကိုတော့အောက်ပါအတိုင်း installation လုပ်လို့ ရပါတယ်။  
```
$ bash Anaconda3-5.2.0-Linux-x86_64.sh
```
4. Welcome message ကိုဖတ်၊ ENTER ကီးနှိပ်ပြီး၊ လိုင်စင်နဲ့ ပတ်သက်တာတွေကိုလည်း page-down လုပ်သွားပါ။  
အားလုံးဖတ်လို့ပြီးရင် "Yes" လို့ရိုက်ထည့်ပေးပါ။ default path အောက်မှာ (ဥပမာ /home/your_login_account_name/anaconda3/) Anaconda ကို installation မလုပ်ချင်ဘူးဆိုရင်၊ ဘယ် path အောက်မှာ install လုပ်မယ်ဆိုတာကို ရိုက်ထည့်ပေးရပါမယ်။  

5. အဆင်ပြေပြေနဲ့ အားလုံး ပြီးစီးသွားမယ်ဆိုရင် အောက်ပါအတိုင်း Anaconda ကို installation လုပ်ထားတဲ့ path ကို ".bashrc" ဖိုင်ရဲ့ PATH environment variable မှာ ဝင်ဖြည့်ပေးမလား ဆိုတာကို မေးပါလိမ့်မယ်။  
```
installation finished.
Do you wish the installer to prepend the Anaconda3 install location
to PATH in your /home/sammy/.bashrc ? [yes|no]
[no] >>> 
```
"yes" လို့ ဖြေပေးပါ။ အဲဒီလို လုပ်ပေးခြင်းအားဖြင့် ဘယ် path အောက်မှာမဆို conda ဆိုတဲ့ command ကို ရိုက်ပြီးတော့  
Anaconda ကို သုံးလို့ ရပါလိမ့်မယ်။  

5. Installation path ကို လက်ရှိသုံးနေတဲ့ terminal မှာ active ဖြစ်အောင် အောက်ပါ command ကို ရိုက်ပါ။  
```
$ source ~/.bashrc
```

6. အထက်ပါ အဆင့်တွေအားလုံးလုပ်ပြီးသွားရင် အောက်ပါအတိုင်း command ပေးလို့ Anaconda ကနေ installation လုပ်ပေးနိုင်တဲ့ package တွေကို list လုပ်ပြတာမြင်ရပြီ ဆိုရင်တော့ Anaconda ကို installation လုပ်တာက အောင်အောင်မြင်မြင်နဲ့ ပြီးဆုံးသွားပြီလို့ ပြောနိုင်ပါတယ်။ 
```
$ conda list
```

## Activate Your New Environment

ကိုယ်ဖန်တီးထားတဲ့ python environment အသစ်ဖြစ်တဲ့ py3.6.5 ထဲကိုဝင်ချင်ရင် အောက်ပါအတိုင်း command ကို ရိုက်ထည့်ပါ။  

```
source activate py3.6.5
```

အောက်ပါအတိုင်း command prompt ရဲ့ အရှေ့ဆုံးနေရာမှာ ကွင်းစကွင်းပိတ်နဲ့ လက်ရှိရောက်နေတဲ့ virtual environment ကို နာမည်နဲ့ တကွ ပြသပေးပါလိမ့်မယ်။  

```
(py3.6.5) lar@lar-air:~$
```

python version ကို check လုပ်ကြည့်လျှင် အောက်ပါအတိုင်း 3.6.5 version ဖြစ်နေတာကို တွေ့ရပါလိမ့်မယ်။  

```
(py3.6.5) lar@lar-air:~$ python --version
Python 3.6.5

```

ကိုယ် run တဲ့ python ကလည်း ဘယ် path အောက်မှာ installation လုပ်ထားတဲ့ python လည်းဆိုတာကို သိချင်ရင်တော့  
အောက်ပါအတိုင်း command ပေးကြည့်ပါ။  

```
(py3.6.5) lar@lar-air:~$ which python
/home/lar/anaconda3/envs/py3.6.5/bin/python
```

အထက်မှာ မြင်ရတာကတော့ ကျွန်တော်ရဲ့ စက်မှာ installation လုပ်ထားတဲ့ Anaconda environment တစ်ခုဖြစ်တဲ့ python 3.6.5 ကို သိမ်းထားတဲ့ path က "/home/lar/anaconda3/envs/py3.6.5/bin/" အောက်မှာ ဆိုတာကို ရိုက်ပြပေးတာ ဖြစ်ပါတယ်။  

<font color='red'> မှတ်ချက်။ ။ "source" ဆိုတဲ့နေရာမှာ "." နဲ့ အစားထိုးလို့ ရပါတယ်။ </font>   

## Deactivate Your Current Enviroment  

လက်ရှိ ဝင်ထားတဲ့ virtual environment ကနေထွက်ချင်ရင်တော့ အောက်ပါအတိုင်း command ကို ပေးပါ။  

```
source deactivate py3.6.5

(or)

source deactivate
```

အဲဒါဆိုရင် အောက်ပါအတိုင်း deactivate လုပ်သွားပြီးလို့ ကွင်းစကွင်းပိတ်နဲ့ ပြနေတဲ့ environment က မရှိတော့မှာကို  
ဂရုပြုပါ။  

```
(py3.6.5) lar@lar-air:~$ source deactivate
lar@lar-air:~$ 
```

## Some Useful Commands of Anaconda

- Anaconda ကို update လုပ်ဖို့ လိုအပ်တယ် ဆိုရင် ``` conda update conda ``` ဆိုပြီး ရိုက်ပါ။  
- ကိုယ့်စက်ထဲမှာ လက်ရှိပြင်ဆင်ထားတဲ့ conda virtual environment တွေကို list လုပ်ပြပေးစေချင်ရင်

$ anaconda-clean

$ rm -rf ~/anaconda3
